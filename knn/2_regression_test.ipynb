```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
# 准备数据
X = [[2,1],[3,1],[1,4],[2,6]]
y = [0.1,9.1,4,3]  # 分类标签
```

```python
# KNN回归模型
knn = KNeighborsRegressor(n_neighbors=2,weights='distance')
# 模型训练
knn.fit(X,y)
# 预测
x = [[4,9]]
x_pred = knn.predict(x)
print(x_pred)
```
