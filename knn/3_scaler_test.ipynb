```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = [[2, 1], [3, 1], [1, 4], [2, 6]]
```

```python
# 定义归一化类的对象
scaler = MinMaxScaler(feature_range=(-1, 1))
# 将缩放器应用到特征上
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
print(X)
```

```python
# 标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

```python
import numpy as np
X = np.array(X)
print(X)
```

```python
mean = np.mean(X, axis=0)
print(mean)  # 平均
std = np.std(X, axis=0)
print(std)  # 方差
# 手动标准化
print('标准化:\n',(X - mean) / (std))  # (X-平均)/方差
```
