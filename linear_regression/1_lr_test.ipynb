```python
from sklearn.linear_model import LinearRegression
```

```python
# 1. 定义数据
# 自变量, 每周学习时间
X = [[5],[8],[10],[12],[15],[3],[7],[9],[14],[6]]
# 因变量, 数学考试成绩
y = [55,65,70,75,85,50,60,72,80,58]
```

```python
# 2. 创建模型: 线性回归
model = LinearRegression()
```

```python
# 3. 模型训练
model.fit(X,y)
```

```python
# 打印模型参数
print(model.coef_)
print(model.intercept_)
```

```python
# 5. 预测
x_new = [[11]]
y_pred = model.predict(x_new)
print(y_pred)
```

```python
# 6. 画图
import matplotlib.pyplot as plot
import numpy as np
```

```python
x_line = np.arange(0,15,0.1).reshape(-1,1)
y_line = model.predict(x_line)
```

```python
plot.scatter(X,y)
plot.plot(x_line,y_line,color='red')
plot.scatter(x_new,y_pred,color='green')
plot.show()
```

```python
# 用方差和协方差验证数学求解公式
x = np.array(X).reshape(-1)
cov = np.cov(x,y)
print(cov)
```

```python
beta = cov[0,1]/cov[0,0]    # 相关系数 = 协方差 / 方差
print(beta)
print(model.coef_)
```

```python
model = LinearRegression(fit_intercept=False)   # 无截距
model.fit(X,y)
```

```python
print(model.intercept_)
print(model.coef_)
```

```python
# 预测
x_new = [[11]]
y_pred = model.predict(x_new)
print(y_pred)
```

```python
# 画图
import matplotlib.pyplot as plot
import numpy as np

x_line = np.arange(0, 15, 0.1).reshape(-1, 1)
y_line = model.predict(x_line)
plot.scatter(X, y)
plot.plot(x_line, y_line, color='red')
plot.scatter(x_new, y_pred, color='green')
plot.show()
```
