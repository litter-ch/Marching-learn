```python
import numpy as np
```

```python
# 随机生成(0,1)均匀分布的数据
a = np.random.rand(5)
print(a)
```

```python
a = np.random.rand(2,3)
print(a)
```

```python
# 随机生成均匀分布的整数
b = np.random.randint(0,100,size=(2,3))
print(b)
```

```python
# 随机生成服从标准正态分布的数据
c = np.random.randn(2,5)
print(c)
```

```python
print((c+5)*10) # 扩大均值, 方差
```
